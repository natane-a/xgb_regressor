```python
X_train = train_data.drop(columns=['price'])
y_train = train_data['price']
X_test = test_data
```

```python
from sklearn.feature_selection import RFE
from xgboost import XGBRegressor
```

```python
model = XGBRegressor(objective='reg:squarederror', n_estimators=100, max_depth=5, learning_rate=0.05)

rfe = RFE(estimator=model, n_features_to_select=10, step=1)

rfe.fit(X_train, y_train)
```

```python
selected_features = X_train.columns[rfe.support_]
print("Selected features:", selected_features.tolist())

feature_ranking = pd.DataFrame({
    'Feature': X_train.columns,
    'Ranking': rfe.ranking_
}).sort_values(by='Ranking')
print(feature_ranking)
```

```python
X_train = train_data.drop(columns=['price', 'hasGuestRoom', 'id', 'isNewBuilt', 'hasPool', 'hasStormProtector', 'hasYard', 'hasStorageRoom'])
y_train = train_data['price']
X_test = test_data.drop(columns=['hasGuestRoom', 'id', 'isNewBuilt', 'hasPool', 'hasStormProtector', 'hasYard', 'hasStorageRoom'])
```

```python
model = XGBRegressor(objective='reg:squarederror', eval_metric='rmsle', n_estimators=1000, learning_rate=0.05)
model.fit(X_train, y_train)

predictions = model.predict(X_test)
X_test['PredictedPrice'] = predictions

print(X_test[['PredictedPrice']].head())
```
